```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
disease = pd.read_csv('heart_disease.csv')
disease
```

```python
disease.info()
```

```python
#getting the numbver of people with heart disease

disease.target.value_counts()
```

```python
#show the percentage of male ove female w
count_female= len(disease[disease.sex==0])
count_male= len(disease[disease.sex==1])

count_female
```

```python
# Calculate the total number of samples
total_samples = len(disease)

# Calculate the percentage of females
percentage_female = (count_female / total_samples) * 100

# Calculate the percentage of males
percentage_male = (count_male / total_samples) * 100

print(f"Percentage of females: {percentage_female:.2f}%")
print(f"Percentage of males: {percentage_male:.2f}%")
```
