```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
cup = pd.read_csv("AfricaCupofNationsMatches.csv")
cup
```

```python

sns.scatterplot(x='HomeTeamGoals', y='Attendance', data=cup)
```

```python
sns.rugplot(x="Attendance", data=cup, height=0.5)
```

```python
sns.set(style='dark')
sns.displot(data=cup, x='Attendance')
```

```python
sns.kdeplot(data=cup, x="Attendance")
```

```python
sns.kdeplot(data=cup, x="Attendance", clip =[0,1000])
```

```python
cup.head()
```

```python

plt.figure(figsize=(10,4),dpi=200)
sns.countplot(data=cup,x="Stage")
```

```python
plt.figure(figsize=(10,4),dpi=200)
sns.barplot(data=cup, x='AwayTeam', y='Attendance', estimator=np.mean,ci='sd')
```
